```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

pd.set_option("display.max.columns", None)
pd.set_option("display.max.rows", None)
```

```python
train = pd.read_csv("data/clean_training.csv")
```

```python
train.head()
```

```python
train.groupby("MSSubClass").mean()
```

```python
sns.set_style("whitegrid")
sns.set_palette("Blues", n_colors=30)

data_p1 = train.groupby("MSSubClass").mean()["SalePrice"]

f, ax = plt.subplots(3, 2, figsize=(14, 21))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.3)
f.suptitle("Different Visualizations for Dataset Exploration", size=20, y=0.95)

# plot 1
sns.barplot(ax=ax[0, 0], x=data_p1.index, y=data_p1)
ax[0, 0].set_title("MSSubClass vs. Average Sale Price", size=16)

# plot 2
sns.histplot(ax=ax[0, 1], data=train, x="GarageArea", hue="GarageCars", palette=sns.color_palette('Blues', n_colors=5),
             kde=True)
ax[0, 1].set_title("Number of Cars vs Garage Area", size=16)

# plot 3
sns.regplot(ax=ax[1, 0], data=train, x="LotArea", y="SalePrice", line_kws={"color": "darkblue"},
            scatter_kws={"alpha": 0.8, "s": 5, "color": "lightblue"})
ax[1, 0].set_title("Lot Area (x) vs. Sale Price (y)", size=16)

# plot 4 
order_idx = train.groupby("Neighborhood").median().sort_values(ascending=True, by="SalePrice").index
sns.boxplot(ax=ax[1,1], x=train["Neighborhood"], y=train["SalePrice"], order=order_idx, linewidth=1, fliersize=1.5)
plt.setp(ax[1, 1].get_xticklabels(), rotation=45, horizontalalignment='right')
ax[1, 1].set_title("Neighborhood (x) vs Sale Prices (y)", size=16)

# plot 5
sns.kdeplot(ax=ax[2,0], x=train["BsmtFinSF1"] + train["BsmtFinSF2"] + train["1stFlrSF"] + train["2ndFlrSF"] 
            + train["LowQualFinSF"], color=sns.color_palette("Blues", n_colors=10)[9], shade=True,
            label="Total Square Feet")
sns.kdeplot(ax=ax[2,0], x=train["TotalBsmtSF"], shade=True, label="Basement Square Feet", color=sns.color_palette("Blues", n_colors=10)[2])
sns.kdeplot(ax=ax[2,0], x=train["1stFlrSF"], shade=True, label="1st Floor Square Feet", color=sns.color_palette("Blues", n_colors=10)[4])
sns.kdeplot(ax=ax[2,0], x=train["2ndFlrSF"], shade=True, label="2nd Floor Square Feet", color=sns.color_palette("Blues", n_colors=10)[7])
ax[2,0].legend()
ax[2, 0].set_title("Square Feet Distribution decomposed", size=16)

# plot 6
sns.kdeplot(ax=ax[2,1], data=train, x="OverallQual", y="OverallCond", cmap="Blues", bw_method=0.25)
ax[2, 1].set_title("Distribution of Quality and Condition", size=16)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 13))
sns.heatmap(ax=ax, data=train.iloc[:,:].corr().abs(), cmap="Blues", linewidths=0.5)
f.suptitle("Correlations of individual features", size=20)
plt.show()
```
